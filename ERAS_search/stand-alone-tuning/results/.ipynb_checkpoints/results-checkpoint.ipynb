```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# draw plot 
def read_result_single(txt_path):
    results = []
    result_epoch, result_valid_mrr, result_test_mrr = [], [], []
    result_file = open(txt_path)
    for line in result_file:
        l = line.split("\t")
        if len(l) ==4:
            result_epoch.append(int(l[0]))
            result_valid_mrr.append(float(l[2].split("  ")[0]))
            result_test_mrr.append(float(l[3].split(" ")[0]))
    
    results.append(result_epoch)
    results.append(result_valid_mrr)
    results.append(result_test_mrr)
    return results


fixed_distmul_path = "WN18RR/WN18RR_fixed_dist.txt"
fixed_analogy_path = "WN18RR/WN18RR_fixed_analogy.txt"
#single_complex_path = "WN18RR/WN18RR_evaluate_complex.txt"

fixed_distmul = read_result_single(fixed_distmul_path)
fixed_analogy = read_result_single(fixed_analogy_path)
#single_complex = read_result_single(single_complex_path)
```

```python
oas_5_file = "WN18RR/WN18RR_oas_kge_5.txt"
oas_7_file = "WN18RR/WN18RR_oas_kge_7.txt"
oas_15_file = "WN18RR/WN18RR_oas_kge_15.txt"

oas_5 = read_result_single(oas_5_file)
oas_7 = read_result_single(oas_7_file)
oas_15 = read_result_single(oas_15_file)
```

```python
def read_result_single(txt_path):
    results = []
    result_epoch, result_valid_mrr, result_test_mrr = [], [], []
    result_file = open(txt_path)
    for line in result_file:
        l = line.split("\t")
        if len(l) ==4 and l[0].split(":")[0] != "best_performance":
                result_epoch.append(int(l[0]))
                result_valid_mrr.append(float(l[2].split("  ")[0]))
                result_test_mrr.append(float(l[3].split(" ")[0]))
    
    results.append(result_epoch)
    results.append(result_valid_mrr)
    results.append(result_test_mrr)
    return results
```
